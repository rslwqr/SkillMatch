```python
from pathlib import Path
import pandas as pd

RAW_DIR = Path("../data/raw")

JOBS_PATH = RAW_DIR / "linkedin_job_postings.csv"
SUMMARY_PATH = RAW_DIR / "job_summary.csv"
SKILLS_PATH = RAW_DIR / "job_skills.csv"
```

```python
jobs_sample = pd.read_csv(JOBS_PATH, nrows=5)
summary_sample = pd.read_csv(SUMMARY_PATH, nrows=5)
skills_sample = pd.read_csv(SKILLS_PATH, nrows=5)
```

```python
print("Jobs columns:")
print(jobs_sample.columns.tolist())

print("\nSummary columns:")
print(summary_sample.columns.tolist())

print("\nSkills columns:")
print(skills_sample.columns.tolist())
```

```python
def count_rows(path):
    with open(path, "rb") as f:
        return sum(1 for _ in f) - 1

print("linkedin_job_postings:", count_rows(JOBS_PATH))
print("job_summary:", count_rows(SUMMARY_PATH))
print("job_skills:", count_rows(SKILLS_PATH))
```
